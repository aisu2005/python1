```python
#rectangle
class Rectangle:
    def __init__(self, l, w):
        self.length = l
        self.width = w
        self.area = l * w

    def __lt__(self, other): 
        if self.area < other.area:
            print("Rectangle 2 is greater")
        else:
            print("Rectangle 1 is greater")
l1 = int(input("Enter length of rectangle 1: "))
a1 = int(input("Enter width of rectangle 1: "))
l2 = int(input("Enter length of rectangle 2: "))
a2 = int(input("Enter width of rectangle 2: "))
r1 = Rectangle(l1, a1)
r2 = Rectangle(l2, a2)
r1 < r2
```

```python
#time overloading
class time:
    def __init__(self,hr,min,sec):
        self.hr=hr
        self.min=min
        self.sec=sec
    def __add__(self,other):
        s=self.sec+other.sec
        m=self.min+other.min
        h=self.hr+other.hr
        if s>=60:
            s=s%60
            m=m+1
        if m>=60:
            m=m%60
            h=h+1
        print("Added time=")
        print(h,"Hour",m,"Minutes",s,"Seconds")
        
hr1=int(input("Enter 1st hour:"))
min1=int(input("Enter 1st minute:"))
sec1=int(input("Enter 1st second:"))
print("------------------")
hr2=int(input("Enter 2nd hour:"))
min2=int(input("Enter 2nd minute:"))
sec2=int(input("Enter 2nd second:"))

t1=time(hr1,min1,sec1)
t2=time(hr2,min2,sec2)
t3=t1+t2        
```
