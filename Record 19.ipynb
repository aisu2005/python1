```python
#2 copy odd lines of one file to another
#input file names from user
file1=input("Enter the source file:")
file2=input("Enter the destination file:")

#open source and destination files
f1=open(file1,"r")
f2=open(file2,"w")

i=0
for line in f1.readlines():
    i=i+1
    if i%2==0: #skip even lines
        continue
    f2.write(line)

#close both files
f1.close()
f2.close()
print("Odd lines copied successfully from",file1,"to",file2)
```

```python
#3 read each row from a csv file and print aslist of strings
import csv
fname = input("Enter CSV file name: ")
f1 = open(fname, "r")
reader = csv.reader(f1)
for row in reader:
    print(row)
f1.close()
```

```python
#4 Read specific columns(DepId,DeptName) from csv file
import csv
fname = input("Enter the CSV file name: ")
with open(fname, "r") as f1:
    reader = csv.DictReader(f1)
    print("\nDepartment Details:")
    print("-------------------")
    for row in reader:
        print("DeptId:", row['DeptId'], "\tDeptName:", row['DeptName'])
    
```

```python
# 5 write a python dictionary to a csv file.after write csv file read it, display the content
import csv
data=[
    {"Name":"Resmi","Age":"22","Marks":"85"},
    {"Name":"Arun","Age":"21","Marks":"78"},
    {"Name":"Meera","Age":"23","Marks":"91"} 
]
with open("Students_dict.csv","w",newline='')as file:
    fieldnames=["Name","Age","Marks"]
    writer=csv.DictWriter(file,fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)
print("Dictionary data written successfully!")
with open("Students_dict.csv","r")as file:
    reader=csv.DictReader(file)
    for row in reader:
        print(row)
```
